```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
def correlacion_pearson(data_frame, columna1, columna2):
    return np.corrcoef(data_frame[columna1], data_frame[columna2])[0,1];
```

# DataSet Heuristicas 1 y 2 


Leemos el dataset de los resultados

```python
h1 = pd.read_csv("resultados/Experimento1/h1.csv");
print(h1)
```

```python
h2 = pd.read_csv("resultados/Experimento1/h2.csv");
print(h2)
```

## Impacto

```python
fig = sns.scatterplot(data=h1, x='N', y='Impacto', color="b", label="H1");
fig = sns.scatterplot(data=h2, x='N', y='Impacto', color="c", label="H2");

fig.set_title('Impacto de las heuristicas')
fig.set(xlabel='#Nodos', ylabel='Impacto');
plt.savefig("graficos/h1vsh2IMPACTO");
```

## Tiempo

```python
fig = sns.scatterplot(data=h1, x='N', y='Tiempo', color="b", label="H1");
fig = sns.scatterplot(data=h2, x='N', y='Tiempo', color="c", label="H2");

fig.set_title('Impacto de las heuristicas')
fig.set(xlabel='#Nodos', ylabel='Tiempo en milisegundos');
plt.savefig("graficos/h1vsh2TIEMPO");
```
