# Generador de instancias
En este notebook está el código para generar los sets de instancias que se usan para experimentar.
- Estas instancias van a ser guardadas en la carpeta __instancias__.
- Cada set estará en su propia carpeta y tendrá un archivo _indice.csv_ que contendrá información sobre las instancias.

```python
import random, math
import pandas as pd
import numpy as np
from networkx.generators.random_graphs import erdos_renyi_graph
import itertools
```

```python
def save_instance(dataset, instance_name, S, W, C):
    with open(F"instancias/{dataset}/{instance_name}.txt", "w") as f:
        print(len(S), W, file=f)
        for i in range(len(S)): 
            print(S[i], C[i], file=f, sep=' ', end ='\n')

def save_index(dataset, instances):
    with open(F"instancias/{dataset}/instances.txt", "w") as f:
        for instance in instances: 
            print(instance, file=f)
```

```python
def generadorDeAristas(file, cantAristas, cantNodos):
    aristas = list(itertools.combinations(range(1, cantNodos + 1), 2))
    #random.shuffle(aristas)
    if(len(aristas) < cantAristas):
        print(f"cantAristas = {cantAristas}")
        print(f"aristas = {len(aristas)}")
        print(f"nodos = {cantNodos}")
        print(aristas)
    for _ in range(cantAristas):
        i = aristas.pop()
        word = ''.join(f"{i[0]} {i[1]}")
        file.write(word + '\n')
```

```python
def txt(txtname, nodes, aristasG, aristasH):
    file = open(f'nuestrasInstancias/{txtname}.txt', 'w+')
    file.write(f'{nodes} {aristasG} {aristasH}' + '\n')
    print(f"aristasG = {aristasG}")
    generadorDeAristas(file, aristasG, nodes)
    print(f"aristasH = {aristasH}")
    generadorDeAristas(file, aristasH, nodes)
    file.close()
```
