# experimento 4

```python
M1 = pd.read_csv("resultados/Experimento5/M1.csv");
M2 = pd.read_csv("resultados/Experimento5/M2.csv");
H2 = pd.read_csv("resultados/Experimento5/H2.csv");
print(M2)
```

```python
names = ['baja', 'media', 'alta']
rows = ['Densidad G Baja', 'Densidad G media', 'Densidad G alta']
cols = ['Densidad H Baja', 'Densidad H media', 'Densidad H alta']
f = plt.figure()    
f, axes = plt.subplots(nrows = 3, ncols = 3, sharex=True, sharey=True, figsize=(17,15))
f.suptitle("Comparando las implementaciones Tiempo", fontsize=30)
#fig.suptitle('averSifunca')
index = 1
for i in range(len(names)):
    for j in range(len(names)):
        data = M1[(M1['densidadG'] == names[i]) & (M1['densidadH'] == names[j])].copy();
        dataM2 = M2[(M2['densidadG'] == names[i]) & (M2['densidadH'] == names[j])].copy();
        dataH2 = H2[(M1['densidadG'] == names[i]) & (H2['densidadH'] == names[j])].copy();
        if i == 2 and j == 2:
            axes[i][j].scatter(data.N, data.Tiempo, color='g', label="Metaheuristica 1");
            axes[i][j].scatter(dataM2.N, dataM2.Tiempo, label="Metaheuristica 2");
            axes[i][j].scatter(dataH2.N, dataH2.Tiempo, color='r', label="heuristica 2");
        else:  
            axes[i][j].scatter(data.N, data.Tiempo, color='g');
            axes[i][j].scatter(dataM2.N, dataM2.Tiempo);
            axes[i][j].scatter(dataH2.N, dataH2.Tiempo, color='r');
        index += 1


for ax, col in zip(axes[0], cols):
    ax.set_title(col)

for ax, col in zip(axes[2], rows):
    ax.set_xlabel("cantidad de nodos")

for ax, row in zip(axes[:,0], rows):
    ax.set_ylabel("tiempo(ms)", rotation=90,)

for ax, row in zip(axes[:,2], rows):
    ax.set_ylabel(row, rotation=90, size='large', labelpad=30)
    ax.yaxis.set_label_position("right")
    ax.yaxis.tick_right()
# fig.set_title('Tiempo de ejecucion de metaheuristicas')
# fig.set(xlabel='#Nodos', ylabel='Tiempo en milisegundos');
f.legend()
plt.savefig("graficos/Experimento5Tiempo.png");
```
